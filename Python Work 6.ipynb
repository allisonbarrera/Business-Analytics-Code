## Used data to train a lasso model.  Learned about the importance of train and test data. Worked with looping through the data set to learn from our data.

```python
import pandas as pd
import pandas
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LassoLarsCV
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

alldata = pd.read_csv('finalmaster-ratios.csv')

allvariablenames = list(alldata.columns.values)

#remove not first, but next 7 variables  *************
del allvariablenames[0:8]

#load predictors into dataframe
predictors = alldata[allvariablenames]  

#load target into dataframe
target = alldata['# Purchases']  
                 
# split data into train and test sets, with 30% retained for test

pred_train, pred_test, tar_train, tar_test = train_test_split(predictors, target, test_size=.3, random_state=123)   

#build the lasso model
#Use LassoLarsCV and feed the following parameters in:
#cv = 10 (this performs a 10-fold cross validation, to make sure our results aren't do to random ordering of the data)
#precompute=False (not necessary)
#name this model 'model'.
#fit the new model to the train data
#use.fit()

from sklearn.linear_model import LassoLarsCV
from sklearn.datasets import make_regression

model = LassoLarsCV(cv=10, precompute=False).fit(pred_train, tar_train)


#build coefficent chart

predictors_model=pd.DataFrame(allvariablenames)  #makes predictor names into a dataframe
predictors_model.columns = ['label']  #makes the column 'label'
predictors_model['coeff'] = model.coef_  #creates a new column in the data frame called 'coeff'

for index, row in predictors_model.iterrows():
    if row['coeff'] > 0:
        print(row.values)

#this loop loops through the data frame and whatever predictor has a coefficient grater than 0
        #and then prints the predictor label and its corresponding coefficient
        
#Mean squared error of the training sets
train_error = mean_squared_error(tar_train, model.predict(pred_train))
print('training data MSE')
print(train_error)  #22528.48
#Mean squared error of the test sets

#Mean squared error of the test sets
test_error = mean_squared_error(tar_test, model.predict(pred_test))
print('Test data MSE')
print(test_error)  #41578.28

#R square for the training set
rsquared_train = model.score(pred_train, tar_train)
print('training data R-square')
print(rsquared_train)  #0.2226

#R square for the test set
rsquared_test = model.score(pred_test, tar_test)
print('test data R-square')
print(rsquared_test)  #0.17529

#y intercept
print('y intercept')
print(model.intercept_)   #2.758738
```
