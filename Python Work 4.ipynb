```python
#Allison Barrera

import pandas as pd
import numpy as np
import datetime 
import statsmodels.api as sm


tp = pd.read_csv('Travel Pony Facebook.csv')
print(tp.head(2))

#1a Advertising performance can vary by time. That is, certain times and days tend
 #to be better for advertising impressions. There are no "magic" times and days. 
 #Instead, it tends to vary by platform and product. That is, every brand/product is different. 
 #Facebook is different from Instagram. 
#So, let's take see what works best for TravelPony on Facebook. 
tp['cost per impression'] = tp['Amount Spent (USD)'] / tp['Impressions']
  
#1b First, create the analytics 'cost per impression' by dividing Amount Spent / Impressions. 
#Do this for each ad in your data, and save that as a column. 
#When considering 'cost per impression', what day of the week works best? 
#(What day is it cheapest to generate impressions) 
#What day works worst? (What day is most expensive)
import arrow
SD = tp['Start Date']
dayofweek = []

for row in range(len(tp)):
    date = arrow.get(tp['Start Date'][row], 'M/D/YY')
    dayofweek.append(date.weekday())
    
tp['Week Day']= dayofweek

group = tp.groupby('Week Day')
group.aggregate(np.mean)

#Saturday's work best for creating cost per impression when looking at the data. It has the most reach and is the cheapest to generate. 
 #The worst and most expensive day is on Friday.
    

#2 Next, compute the correlation(Links to an external site.)Links to an external site.
#between Amount Spent and the following variables:
#- Reach
#- Frequency 
#- Unique Clicks
#- Page Likes
 #Which correlation is the strongest? What does that mean practically? (respond in a tweet or less)
AmountReach = np.corrcoef(tp['Amount Spent (USD)'], tp['Reach'])  #0.70312381
AmountFreq = np.corrcoef(tp['Amount Spent (USD)'], tp['Frequency']) #0.13020087
AmountUnique = np.corrcoef(tp['Amount Spent (USD)'], tp['Unique Clicks']) #0.88299318
AmountPageLikes = np.corrcoef(tp['Amount Spent (USD)'], tp['Page Likes']) #0.75761193

#The amount and unique clicks had the strongest correlation which means that it is best to spend money 
#on unique clicks becuase it is the most effective

#3 Finally, perform a simple multiple regression analysis (Links to an external site.)
#Links to an external site.where Unique Clicks is the dependent variable and Reach 
#and Frequency are the independent (predictor) variables.
#What variable most strongly predicts unique clicks? What does that mean practically? 
#(respond in a tweet or less)
y=tp['Unique Clicks']
x=tp[['Reach', 'Frequency']]

model = sm.OLS(y,x).fit()
predictions = model.predict(x)

print (model.summary())


#Frequenecy is the one that is most predictive because it has the highest coeffienent
#and they are both statisically significant
```
